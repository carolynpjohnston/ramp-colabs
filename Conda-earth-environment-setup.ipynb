Next, create a raster for the mask with 1 channel, and the same dimensions and CRS as the building raster. I found a function that does this [here](https://stackoverflow.com/questions/51482423/create-empty-copy-of-raster-with-python-gdal-api).

```python
def make_blank_mask_from_img(ds, mask_path):
    '''
    ds: gdal raster dataset (we'll match its georeferencing and size in the byte mask)
    mask_path: where to write the mask file.
    '''
    dr = ds.GetDriver()
    ds_new = dr.Create(mask_path,ds.RasterXSize,ds.RasterYSize,ds.RasterCount,gdal.GDT_Byte)
    ds_new.SetGeoTransform(ds.GetGeoTransform())
    ds_new.SetProjection(ds.GetProjection())
    return ds_new 
```

```python

# define paths, open the raster source file
raster = r'/content/drive/MyDrive/SN2-Khartoum/pansharp/RGB-PanSharpen_AOI_5_Khartoum_img1.tif'
mask = r'/content/drive/MyDrive/SN2-Khartoum/bldgs-mask/RGB-PanSharpen_AOI_5_Khartoum_mask1.tif'
json = r'/content/drive/MyDrive/SN2-Khartoum/bldgs-geojson/buildings_AOI_5_Khartoum_img1.geojson'
ras_ds = gdal.Open(raster)
```

```python
# WHAT a hassle it was to get this to work. But this did finally work.
mask_ds = make_blank_mask_from_img(ras_ds, mask)
# get vector layer
lyr = None
with open(json) as f:
  the_json = f.read() # I checked: the_json is not empty.
  ds = ogr.Open(the_json)
  lyr = ds.GetLayer()

# Show there are buildings in the layer; 26 of them.
len(lyr)

# burn vector layer into raster mask. Nullify the mask_ds variable at the end, to flush the image to disk.
gdal.RasterizeLayer(mask_ds, [1], lyr) 
mask_ds.GetRasterBand(1).SetNoDataValue(0.0) 
mask_ds = None
```
